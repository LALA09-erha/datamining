## **2. Membuat statistik diskriptif dari data (bertipe numerik)**

**Rumus mean :** <br>
>  𝓶  = $\frac {Σxˌ}{𝓷}$<br>

```
  ket:
  - 𝓶 = mean/ nilai rata - rata 
  - 𝒙ˌ = jumlah data
  - 𝓷  = banyaknya data
```

**Rumus Max :**<br>
>  Nilai Terakhir dari 𝐗[n] ;<br>

```
  ket:
  - 𝐗 = Daftar nilai yang di urutkan dalam set data
  - n = Jumlah nilai dalam set data
```
**Rumus Min :**<br>
>  Nilai Pertama dari 𝐗[n] ;<br>

```
  ket:
  - 𝐗 = Daftar nilai yang di urutkan dalam set data
  - n = Jumlah nilai dalam set data
```
**Rumus Simpangan Baku :**<br>
>  𝞼 =  $\sqrt{\frac{Σ(𝒙ˌ-𝜇)^2}{N}} $

```
  ket:
  - 𝞼  = Simpangan baku populasi
  - N  = Jumlah Populasi
  - 𝔁ᵢ = Setiap nilai dari populasi
  - 𝞵  = rata - rata populasi
```
**Rumus Median :**<br>
> Med(𝑿) = 𝑿[$\frac{n}{2}$] ; Jika n Genap <br><br>
> Med(𝑿) = $\frac{𝑿[\frac{n-1}{2}]+ 𝑿[\frac{n+1}{2}]}{2}$ ; Jika n Ganjil

```
  ket:
  𝑿 = Daftar nilai yang diurutkan dalam set data
  n = Jumlah nilai dalam set data
```

**Rumus Rentang Data / Range :**
> R = 𝐱 max - 𝐱 min

```
  ket:
  R = Rentang data
  x = nilai dari suatu kumpulan data
```

## **3. Mendeskripsikan Skewned dari data setiap variable (bertipe numerik)**

**Rumus SKEW / ukuran kemiringan :** 
> Sk = $\frac{3(𝜇 - Med)}{σ} $

```
  ket:
  Sk  = skew / ukuran kemiringan
  𝜇   = nilai rata - rata hitung
  Med = median
  σ   = simpangan baku / standar deviasi
```

````python
data_num = df.select_dtypes(['int64','float64'])
for colom in data_num:
  print(' ')
  print(colom)
  print(skew(data_num[colom]))
  plt.figure()
  sns.displot(data_num[colom],kde=True)
  plt.title("Skew Data tipe Numerik")
  plt.show()
````

## 1. Lakukan Proses diskritisasi dengan equal width dan equal frequency

### import data

**Rumus Equal Width :**
> 𝓌 = [$\frac{max - min}{𝓍}$]

```
  ket:
  𝓌  = Width
  max = nilai maximum pada data
  min = nilai minimum pada data
  x   = jumlah kategori / bin 
```

**Rumus Equal Frequency :**
> 𝓕req = $\frac{n}{𝓍}$

```
  ket:
  𝓕req = frekuensi
  𝓍    = jumlah kategori / bin
  n    = jumlah data pada list / count
```

````python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import math
from IPython.display import Markdown as md
data = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
data
````

````python
binning_Equal_width = pd.cut(data['sepal.length'],2)
middle_width = binning_Equal_width[0].right
print(" Melakukan proses diskritasi dengan Equal Width  {} < dan <= {}".format(middle_width,middle_width))
````

````python
binning_Equal_frequency = pd.qcut(data['sepal.length'],2)
middle_frequency = binning_Equal_frequency[0].right
print(" Melakukan proses diskritasi dengan Equal Frequency  {} < dan <= {}".format(middle_frequency,middle_frequency))
````

## **2. diskritisasi data panjang sepal dengan equal Width**

### Proses diskritisasi dengan basis entropy

Melakukan proses diskritasi dengan Equal Width pertama dengan memasukkan data pada rumus Entropy

**Rumus Entropy :**
> *Entropy*$(D_1) = - ∑_{i=1}^{m}p_i log_2 p_i$

```
  ket :
  D  = himpunan kasu
  m  = jumlah partisi D
  pi = proporsi dari Di terhadap D
``` 

### **Menghitung Data Pertama Width $(D_1)$**

````python
equal_width1_1 = data[(data['sepal.length']> middle_width )&( data['variety']=='Setosa')].count()[1]
equal_width1_2 = data[(data['sepal.length']> middle_width )&( data['variety']=='Versicolor')].count()[1]
equal_width1_3 = data[(data['sepal.length']> middle_width )&( data['variety']=='Virginica')].count()[1]
jumlah_width1 = equal_width1_1 +equal_width1_2 + equal_width1_3
````

````python
equal_width1_1
````

````python
equal_width1_2
````

````python
equal_width1_3
````

````python
jumlah_width1
````

Pada ***equal_width1_1*** memiliki value 0 dan  $Log_2(0)$ pada bahasa python menggunakan libary math.log2() akan menghasilkan pesan error karena hasilnya undefined maka sebagai alternative pada ***data_width1_1*** akan diisi dengan ***equal_width1_1***

````python
data_width1_1 = equal_width1_1
data_width1_1
````

````python
data_width1_2 = -(equal_width1_2/jumlah_width1)*math.log2(equal_width1_2/jumlah_width1)
data_width1_2
````

````python
data_width1_3 = -(equal_width1_3/jumlah_width1)*math.log2(equal_width1_3/jumlah_width1)
data_width1_3
````

````python
jumlah_data_width1 = data_width1_1 +data_width1_2+data_width1_3
jumlah_data_width1
````

### **Menghitung Data Kedua Width $(D_2)$**

````python
equal_width2_1 = data[(data['sepal.length']<= middle_width )&( data['variety']=='Setosa')].count()[1]
equal_width2_2 = data[(data['sepal.length']<= middle_width )&( data['variety']=='Versicolor')].count()[1]
equal_width2_3 =  data[(data['sepal.length']<= middle_width )&( data['variety']=='Virginica')].count()[1]
````

````python
equal_width2_1
````

````python
equal_width2_2
````

````python
equal_width2_3
````

````python
jumlah_width2 = equal_width2_1+equal_width2_2+equal_width2_3
jumlah_width2
````

````python
data_width2_1 = -(equal_width2_1/jumlah_width2)*math.log2(equal_width2_1/jumlah_width2)
data_width2_1
````

````python
data_width2_2 = -(equal_width2_2/jumlah_width2)*math.log2(equal_width2_2/jumlah_width2)
data_width2_2
````

````python
data_width2_3 = -(equal_width2_3/jumlah_width2)*math.log2(equal_width2_3/jumlah_width2)
data_width2_3
````

````python
jumlah_data_width2 = data_width2_1 + data_width2_2 + data_width2_3
jumlah_data_width2
````

### **Menghitung Data Keseluruhan Width (D)**

Untuk menghitung data keseluruhan, juga memakai rumus Entropy

````python
jumlahkeselurusah_width = jumlah_width1 +jumlah_width2
jumlahkeselurusah_width
````

````python
data_width_1 = -(jumlah_width1/jumlahkeselurusah_width)*math.log2(jumlah_width1/jumlahkeselurusah_width)
data_width_2 =- (jumlah_width2/jumlahkeselurusah_width)*math.log2(jumlah_width2/jumlahkeselurusah_width)
jumlah_data_width = data_width_1 +data_width_2
jumlah_data_width
````

### **Menghitung Data Width $info_a$**

**Rumus :**
> $info_a = \frac{|D_1|}{|D|} Entropy(D_1) + \frac{|D_2|}{|D|} Entropy(D_2)$
```
  ket:
  D1/D2      = jumlah data
  D          = jumlah keseluruhan data
  Entropy(D) = hasil entropy data 
```

````python
infoa_width = ((jumlah_width1/jumlahkeselurusah_width)*jumlah_data_width1) + ((jumlah_width2/jumlahkeselurusah_width) *jumlah_data_width2)
infoa_width
````

### **Menghitung Data Width Gain $(D_{new)}$**

**Rumus Gain$(D_1)$ :**
> $Gain(E_{new}) = E_{initial} - E_{new}$

```
  ket:
  Gain     = gain
  Einitial = jumlah_data_width
  Enew     = infoa
```

````python
gain_width = jumlah_data_width - infoa_width
gain_width
````

### **Tranformasi**

untuk tranformasi, data yang kurang dari **middle_width** akan di kelompokkan menjadi kelas `Setosa`, lalu untuk yang sama dengan **middle_width** akan di kelompokkan menjadi kelas `VersiColor`,sedangkan untuk yang lebih dari middle_width akan di kelompokkan menjadi kelas `Virginica`

````python
data_sepal_width = []
for i in data['sepal.length']:
  if(i < middle_width):
    data_sepal_width.append([i,'Setosa'])
  elif(i==middle_width):
    data_sepal_width.append([i,'Versicolor'])
  elif(i> middle_width):
    data_sepal_width.append([i,'Virginica'])
tf = pd.DataFrame(data=data_sepal_width,columns=['sepal.length','kelas'])
tf
````
